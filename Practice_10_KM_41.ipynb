### **Завдання 1 (1 бал)**

Керівництво будівельної компанії вирішило підняти заробітні плати всім своїм працівникам на 30%. В залежності від спеціалізації та стажу працівники мають різну заробітну плату. Для підготовки звіту необхідно розрахувати нову заробітну плату для працівників компанії, а також суму індексації.

Написати програму, яка буде генерувати таблицю для цього звіту. Початкова заробітна плата працівників подана у списку salary_list (у тисячах гривень). 

1. Для підрахунку невідомих значень повинні використовуватись вбудовані функції вищих порядків.

2. Зарплата після індексації та сума індексації для працівників різних категорій повинні зберігатись у двох окремих списках. Порядок значень у кожному з нових списків повинен відповідати порядку значень у списку salary_list.

3. Кожне підраховане числове значення має бути округлене до сотих.

4. Дані повинні виводитись у вигляді:

**Salary table:**

**6.4 8.32 1.92**

**9.35 12.16 2.81**

**...**

salary_list = [6.4, 9.35, 11.4, 14, 23.8, 28.15, 34.7]

```python
salary_list = [6.4, 9.35, 11.4, 14, 23.8, 28.15, 34.7]

def index(salary_list):
    return round(salary_list * 0.30 + 0.001, 2)
def final_sl(salary_list):
    return round(salary_list + salary_list * 0.30 + 0.001, 2)

index_2 = list(map(index, salary_list))
final_sl_2 = list(map(final_sl, salary_list))

print('Salary table:')
i = 0
for id in range(7):
    print(f'{salary_list[i]} {final_sl_2[i]} {index_2[i]}')
    i += 1
```

Зазвичай рік має 365 днів. Однак у високосні роки додається ще один день - 29 лютого. Правила визначення високосного року такі:

- Будь-який рік, що ділиться на 400 - високосний.

- З тих років, що залишились, будь-який рік, що ділиться на 100 - **не** високосний.

- З тих років, що залишились, будь-який рік, що ділиться на 4 - високосний.

- Усі інші роки, що залишились - **не** високосні.

1) Створіть функцію, в якій за допомогою вбудованих функцій вищих порядків будуть визначатися високосні роки серед перелічених років у списку years. Створена функція має повертати список високосних років.

2) Створіть функцію вищих порядків, яка прийматиме на вхід попередню функцію та визначатиме кількість днів у місяці. Також функція повинна приймати на вхід ще 2 параметри: ціле число від 1 до 12 (місяць) та рік, як ціле чотирицифрове число. Ці значення корисувач має вводити з клавіатури. Перевірку введених даних слід виконувати поза межами функцій.

```python
import numpy as np

years = np.arange(1900, 2020+1, 1)

def vys_y (m):
    return m % 400 == 0 or (m % 100 != 0  and m % 4 == 0)

def fn_vys_y (years):
    return list(filter(vys_y, years))

def fn_days (m, n):
    t = fn_vys_y(years)
    if n == 2:
        return 29 if m in t else 28
    elif n == 4 or n == 6 or n == 9 or n == 11:
        return 30
    else:
        return 31

def main ():
    try:
        m = int(input('Введіть рік: '))
        n = int(input('Введіть місяць (від 1 до 12): '))
        
        if n < 1 or n > 12:
            raise ValueError ('Потібно вводити цілі числа від "1" до "12"')
        
        a = fn_vys_y(years)
        print('Високосні роки:', ','.join(map(str, a)))
        
        w = fn_days(m, n)
        print(f'В {n}-ому місяці {m} року, є {w} день/ів.')
        
    except ValueError as v:
        print('Помилка:', v)
        
if __name__ == '__main__':
    main()
```
